```python
import re

results = {}
for injection_type in ['contextual', 'structural', "dice", "path", "cont_struc", "path_dice"]:
    for sens_rate in [0, 0.5, 1]:
        file_name = f'adone_{injection_type}_flip_{sens_rate}.log'
        
        with open(file_name, 'r') as file:
            content = file.read()
            
            if injection_type not in results:
                results[injection_type] = {}
            if sens_rate not in results[injection_type]:
                results[injection_type][sens_rate] = {}
            
            matches = re.findall(r'Trial\s+(\d+)\s+:.*?({[^}]+})\s*Trial\s+\d+\s+cf:\s*({[^}]+})', content, re.DOTALL)
            
            if matches:
                for match in matches:
                    trial_number = match[0]
                    metrics = eval(match[1])
                    cf_metrics = eval(match[2])
                    results[injection_type][sens_rate][trial_number] = {
                        'metrics': metrics,
                        'cf_metrics': cf_metrics
                    }
                    if match == matches[-1]:
                        best_value = metrics.get('auc', 'N/A')
                        print(f"{injection_type}, {sens_rate}, auc: {best_value},  cf:{metrics['cf']}")
            else:
                print(f"No match found in {file_name}")

print(results)
```
